```python
# 3. Load geographic data
gdf = gpd.read_file("https://raw.githubusercontent.com/rowanhogan/australian-states/master/states.geojson")

print("Estados no arquivo geográfico:")
print(gdf["STATE_NAME"].tolist())
print()
```

```python
# 4. Data
dados = pd.DataFrame(df_copy)
print("Dados originais:")
print(dados)
print()
```

```python
print(gdf)
```

```python
# 5. Mapping and creating the 'State' column
mapping = {
    "New South Wales": "NSW",
    "Victoria": "Vic.",
    "Queensland": "Qld",
    "South Australia": "SA",
    "Western Australia": "WA",
    "Tasmania": "Tas.",
    "Northern Territory": "NT",
    "Australian Capital Territory": "ACT"
}
```

```python
# 6. Checking mapping
gdf["State"] = gdf["STATE_NAME"].map(mapping)

print("Mapping created:")
print(gdf[["STATE_NAME", "State"]].head(10))
print()
```

```python
# 7. Merge using 'State' column
gdf_merged = gdf.merge(dados, on="State", how="left")
```

```python
gdf_merged
```

```python
gdf_merged['2006'][6] = '79.0'
gdf_merged['2011'][6] = '72.0'
gdf_merged['2016'][6] = '60.0'
gdf_merged['2021'][6] = '56.0'
```

```python
gdf_merged['2006'] = ['34.0', '35.0', '48.0', '37.0', '42.0', '24.0', '79.0', '30.0']
```

```python
gdf_merged
```

```python
# 8. Plot
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
gdf_merged.plot(
    column="2006", 
    cmap="GnBu", alpha=0.7,
    linewidth=0.8, 
    #color="lightgrey",
    #edgecolor="0.8", 
    edgecolor="black",
    legend=True,
    ax=ax,
    missing_kwds={"color": "lightgrey", "label": "No data"}  # Estados sem dados em cinza
 )

# Labels
for idx, row in gdf_merged.iterrows():
    # Calculate the centroid (approximate center) of the polygon to position the label
    centroid = row['geometry'].centroid
    ax.annotate(text=row['STATE_NAME'], xy=(centroid.x, centroid.y), 
                xytext=(0, 0), textcoords='offset points', ha='center', va='center',
                fontsize=8, color='black', fontweight='bold')
    
plt.title("Homelessness rate per 10,000 people (2006)", fontsize=16, pad=20)
plt.axis("off")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
# Plot com tratamento de valores nulos
gdf_merged.plot(
    column="2011", 
    cmap="GnBu", alpha=0.7,
    linewidth=0.8, 
    #color="lightgrey",
    #edgecolor="0.8", 
    edgecolor="black",
    legend=True,
    ax=ax,
    missing_kwds={"color": "lightgrey", "label": "No data"}  # Estados sem dados em cinza
 )

# Adicionar rótulos dos nomes dos estados
for idx, row in gdf_merged.iterrows():
    # Calcular o centróide (centro aproximado) do polígono para posicionar o rótulo
    centroid = row['geometry'].centroid
    ax.annotate(text=row['STATE_NAME'], xy=(centroid.x, centroid.y), 
                xytext=(0, 0), textcoords='offset points', ha='center', va='center',
                fontsize=8, color='black', fontweight='bold')
    
plt.title("Homelessness rate per 10,000 people (2011)", fontsize=16, pad=20)
plt.axis("off")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
gdf_merged.plot(
    column="2016", 
    cmap="GnBu", alpha=0.7,
    linewidth=0.8, 
    #color="lightgrey",
    #edgecolor="0.8", 
    edgecolor="black",
    legend=True,
    ax=ax,
    missing_kwds={"color": "lightgrey", "label": "No data"}  # Estados sem dados em cinza
 )

for idx, row in gdf_merged.iterrows():
    
    centroid = row['geometry'].centroid
    ax.annotate(text=row['STATE_NAME'], xy=(centroid.x, centroid.y), 
                xytext=(0, 0), textcoords='offset points', ha='center', va='center',
                fontsize=8, color='black', fontweight='bold')
    
plt.title("Homelessness rate per 10,000 people (2016)", fontsize=16, pad=20)
plt.axis("off")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
gdf_merged.plot(
    column="2021", 
    cmap="GnBu", alpha=0.7,
    linewidth=0.8, 
    #color="lightgrey",
    #edgecolor="0.8", 
    edgecolor="black",
    legend=True,
    ax=ax,
    missing_kwds={"color": "lightgrey", "label": "No data"}  # Estados sem dados em cinza
 )

for idx, row in gdf_merged.iterrows():
   
    centroid = row['geometry'].centroid
    ax.annotate(text=row['STATE_NAME'], xy=(centroid.x, centroid.y), 
                xytext=(0, 0), textcoords='offset points', ha='center', va='center',
                fontsize=8, color='black', fontweight='bold')
    
plt.title("Homelessness rate per 10,000 people (2021)", fontsize=16, pad=20)
plt.axis("off")
```
